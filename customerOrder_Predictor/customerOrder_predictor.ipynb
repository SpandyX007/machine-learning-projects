## Now that we have removed the null values, we can train a model

## Let's use Random forest CLassifier and LinearSVC

```python
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X=df.drop("PurchaseStatus",axis=1)
y=df["PurchaseStatus"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

modelSVC=LinearSVC()
modelRFC=RandomForestClassifier(n_jobs=-1)
```

```python
# acheiving the raw scores in both the models
modelSVC.fit(X_train,y_train)
modelRFC.fit(X_train,y_train)
SVCscore=modelSVC.score(X_test,y_test)
RFCscore=modelRFC.score(X_test,y_test)

print(f"SVC score={SVCscore*100:.2f}%\nRFC score={RFCscore*100:.2f}%")
```

## Since RandomForestClassifier has the most score, will go forward with the RFC model

```python
# average score with cross_val_score
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(modelRFC,X,y,cv=5)
np.mean(cvs)
```

```python
# applying grid search cv to hypertune the model
%time
from sklearn.model_selection import RandomizedSearchCV
np.random.seed(42)
params={"n_estimators":np.arange(10,100,10),
        "min_samples_split":np.arange(2,50,2),
        "min_samples_leaf":np.arange(1,50,2),
        "max_depth":[None,3,5,10,15]}

RFChytu=RandomizedSearchCV(RandomForestClassifier(n_jobs=-1),param_distributions=params, n_iter=1000,verbose=True,cv=5)
RFChytu.fit(X,y)
```

```python
RFChytu.best_params_
```

```python
finalmodel=RandomForestClassifier(n_estimators=10,n_jobs=-1,min_samples_leaf=45, min_samples_split=10, max_depth=None, random_state=42)
finalmodel.fit(X_train,y_train)
finalmodel.score(X_test,y_test)
```

```python
# average score with cross_val_score
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(finalmodel,X,y,cv=5)
np.mean(cvs)
```

## Though we got a reduction in  the R^2 score on the X_train and y_train data, we have got a better cross validation score for our model
